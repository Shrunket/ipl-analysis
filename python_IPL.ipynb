# Highest Strike Rate (16-20 ovr)

```python
pro= deliveries.copy()
```

```python
ovr_17= pro['over']== 17
ovr_18= pro['over']== 18
ovr_19= pro['over']== 19
ovr_20= pro['over']== 20
```

```python
pro=pro[ovr_17 | ovr_18 | ovr_19 | ovr_20]
```

```python
runs= pro.groupby(['batsman'])['batsman_runs'].agg('sum')
```

```python
runs=pd.DataFrame(runs)
```

```python
ball= pro.groupby(['batsman'])['ball'].agg('count')
```

```python
ball=pd.DataFrame(ball)
```

```python
runs['ball']= ball['ball'].values
```

```python
runs['strike_rate']= np.nan
```

```python
for i in range(0,len(runs['strike_rate'])):
    runs['strike_rate'][i]= round(((runs['batsman_runs'][i]*100) / runs['ball'][i]),1)
```

```python
high_Sr= runs[runs['batsman_runs']>=500].sort_values(by='strike_rate', ascending= False).head(10)
```

```python
high_Sr.drop(['batsman_runs','ball'],axis=1, inplace=True)
```
